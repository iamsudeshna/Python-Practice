```python
import pandas as pd
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alexa', 'Amaya', 'Alminas', 'Alicia', 'Ayoub'], 
        'last_name': ['Atherton', 'Sujaya', 'Dolas', 'Young', 'Khan']}


df1 = pd.DataFrame.from_dict(raw_data_1)
df1
```

```python
raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Bijoy', 'Brijesh', 'Bran', 'Bharat', 'Betsy'], 
        'last_name': ['Bondi', 'Bhardwaj', 'Bruce', 'Bisht', 'Binny']}


df2 = pd.DataFrame.from_dict(raw_data_2)
df2
```

```python
#b.	Stack dataframes df1 and df2 by rows (Hint: You donâ€™t have to merge but concatenate). Call new dataframe as df1_df2_rows
df1_df2_rows = df1.append(df2)
df1_df2_rows = df1_df2_rows.reset_index(drop=True)
df1_df2_rows
```

```python
#c.	Stack dataframes df1 and df2 by columns. Call new dataframe as df1_df2_cols
df1_df2_cols = df1 + df2
df1_df2_cols = df1_df2_cols.reset_index(drop=True)
df1_df2_cols
```

```python
#d.	Merge df1_df2_rows with df3 based on subject_id. 
raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

df3 = pd.DataFrame.from_dict(raw_data_3)

df_rows_merged = df1_df2_rows.merge(df3, on='subject_id', how='left')
df_rows_merged
```

```python
#e.	Perform a full join on df1 and df2 based on subject_id.
df1_merged_df2 = pd.merge(df1, df2, on='subject_id', how ='left')
df1_merged_df2
```
